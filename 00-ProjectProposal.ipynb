## Research Question

How do student stress levels vary throughout the academic term at university campuses, and does this variation differ between quarter-system and semester-system institutions? To what extent can academic calendar structure (quarter vs. semester) predict student stress levels over the course of an academic term?

## Background and Prior Work

Academic demands are one of the most significant contributors to university students having consistently high levels of stress. Stress among college students has been linked to harmful outcomes such as reduced academic performance, mental health challenges, and even sleep disruption. While research on stress levels among students has been popular and important, there has been a shift towards understanding how stress can fluctuate over time within an academic term. These patterns matter because stress is not static, it rises and falls in response to pressures like deadlines and exams. Understanding when stress peaks can inform institutional policies and academic support services. 

Prior research that tracks students over time generally finds that stress rises during high demand periods, especially around midterms and finals, rather than remaining constant throughout a whole term. An example of this is a longitudinal study of incoming college students that reported higher average stress during examination periods which linked higher stress to poor sleep quality and negative emotions.
https://pmc.ncbi.nlm.nih.gov/articles/PMC6122850/ The last decade has seen an increase in the use of higher-frequency measurement designs that are suited to detecting within term fluctuations. Research using this approach argues that near real time sampling can capture short-lived spikes and recoveries.This provides a template for aligning stress measurements with specific academic events. This shift motivates data research that models stress as a time varying outcome rather than a single static stress level. 

The last decade has also seen growth in research connecting academic structure to student well-being outcomes. Studies on workload distribution suggest that when academic demands are compressed into shorter time frames, students report greater time pressure and fewer opportunities to recover between assessments. This indicates that not only the amount of academic work, but the pacing of the work, influences how stress develops across a term. Because of this, the organization of an academic calendar may shape when stress begins to rise and how quickly it intensifies during the semester or quarter.
Comparisons between quarter and semester systems provide a useful framework for examining these differences. Quarter systems typically involve faster content coverage and more frequent transitions between courses while semester systems allow longer adjustment periods and a more extended timeline for assignments and exams. Although both systems include predictable stress points during midterms and finals, the compressed structure of the quarter system may lead to earlier spikes in stress.

## Hypothesis


We hypothesize that students in quarter-system universities will experience higher stress earlier in the academic term compared to students in semester-system universities. Additionally, we expect student stress levels to increase more rapidly over time in quarter-system institutions, reflecting the compressed academic pacing as opposed to the slower structure of semester systems.. Finally, we hypothesize that the emphasis of midterms and finals on quarter systems will partially mediate the relationship between academic system structure and student stress.
